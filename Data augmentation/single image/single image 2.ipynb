```python
datagen=ImageDataGenerator(
      rotation_range=30,       # Rotate image randomly by 30 degrees
    width_shift_range=0.2,   # Shift width by 20%
    height_shift_range=0.2,  # Shift height by 20%
    shear_range=0.2,         # Shear transformation
    zoom_range=0.2,          # Zoom in/out by 20%
    horizontal_flip=True,    # Flip image horizontally
    fill_mode='nearest' 
)
datagen
```

```python
output_directory=r"C:\Users\wwrao\Documents\deep learning\Data augmentation\image augmentated"
```

```python
Image=Image.reshape(1,200,200,3)
```

```python
Image.shape
```

```python
aug_iter=datagen.flow(Image,save_to_dir=output_directory,
             batch_size=1,
             save_format='png', save_prefix='aug',)
```

```python
for i in range(100):  # Loop 10 times
    next(aug_iter)
```
