```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen=ImageDataGenerator(
     rotation_range=30,       # Rotate image randomly by 30 degrees
    width_shift_range=0.2,   # Shift width by 20%
    height_shift_range=0.2,  # Shift height by 20%
    shear_range=0.2,         # Shear transformation
    zoom_range=0.2,          # Zoom in/out by 20%
    horizontal_flip=True,    # Flip image horizontally
    fill_mode='nearest'      
)
datagen
```

```python
image.shape
```

```python
batch_image=image.reshape(1,200,200,3)
```

```python
batch_image.shape
```

```python
output_directory=r"C:\Users\wwrao\Documents\deep learning\Data augmentation\new image"
```

```python
import os
```

```python
os.makedirs(output_directory,exist_ok=True)
```

```python
aug_iter = datagen.flow(
    batch_image,
    batch_size=1,
    save_to_dir=output_directory,
    save_prefix='aug',
    save_format='png'
)
```

```python
for i in range(10):  # Loop 10 times
    next(aug_iter)

        
```
