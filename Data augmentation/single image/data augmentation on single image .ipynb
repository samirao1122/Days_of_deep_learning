```python
from tensorflow.keras.preprocessing import image 
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen=ImageDataGenerator(
      rotation_range=30,       # Rotate image randomly by 30 degrees
    width_shift_range=0.2,   # Shift width by 20%
    height_shift_range=0.2,  # Shift height by 20%
    shear_range=0.2,         # Shear transformation
    zoom_range=0.2,          # Zoom in/out by 20%
    horizontal_flip=True,    # Flip image horizontally
    fill_mode='nearest' 
)
```

```python
img=image.load_img(r"E:\Dogs cat dataset\dogs_vs_cats\test\dogs\dog.1884.jpg",target_size=(200,200))
```

```python
image_array=image.img_to_array(img)
```

```python
image_array
```

```python
Image=image_array.reshape(1,200,200,3)
```

```python
Image.shape
```

```python
output_directory=r"C:\Users\wwrao\Documents\deep learning\Data augmentation\image augmentated"
```

```python
counter = 0 
for batch in datagen.flow(Image, save_to_dir=output_directory, 
                          batch_size=1, save_format='png', save_prefix='aug'):
    
    counter += 1  
    if counter == 20:  
        break
```
