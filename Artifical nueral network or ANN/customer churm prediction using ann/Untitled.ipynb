```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"C:\Users\wwrao\Downloads\archive (1)\loan_data.csv")
```

```python
dataset.head()
```

```python
dataset=dataset.drop(columns=["inq.last.6mths","delinq.2yrs","pub.rec","not.fully.paid"])
```

```python
dataset.head()
```

```python
x=dataset.drop(columns=["credit.policy"])
```

```python
y=dataset["credit.policy"]
```

```python
x.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ss=OneHotEncoder()
```

```python
vector=ss.fit_transform(x[["purpose"]]).toarray()
```

```python
col=dataset["purpose"].unique()
```

```python
d1=pd.DataFrame(vector,columns=col)
```

```python
x=pd.concat([x,d1],axis=1)
```

```python
x.drop(columns=["purpose"],inplace=True)
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
import tensorflow as tf
```

```python
from tensorflow import keras 
```

```python
model=tf.keras.Sequential([
    keras.layers.Dense(15,input_shape=(15,),activation="sigmoid"),
    keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
x_train.shape,y_train.shape
```

```python
history=model.fit(x_train,y_train,epochs=500)
```

```python
y_pred=model.predict(x_test)
```

```python
threshold = 0.5
y_predict = []

for i in range(len(y_pred)): 
    if y_pred[i] <= threshold:
        y_predict.append(0)
    else:  
        y_predict.append(1)

# Return the predictions
print(y_predict)  
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```
